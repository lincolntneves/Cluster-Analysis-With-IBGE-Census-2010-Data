```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Suponha que X seja a base j√° normalizada e preparada
X = df_pessoas_v02_transf_02_norm_amostra_01.iloc[:, 1:]

# Lista de diferentes n√∫meros de clusters para testar
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9]

# Dicion√°rio para armazenar os Silhouette Scores
silhouette_scores = {}

for n_clusters in range_n_clusters:
    # Inicialize e ajuste o modelo KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)

    # Calcule o Silhouette Score
    silhouette_avg = silhouette_score(X, cluster_labels)

    # Armazene o Silhouette Score no dicion√°rio
    silhouette_scores[n_clusters] = silhouette_avg

    print(f"Para n_clusters = {n_clusters}, o Silhouette Score m√©dio √© : {silhouette_avg:.3f}")

# Exibindo os Silhouette Scores calculados
silhouette_scores
```

```python
import pandas as pd
import numpy as np

# Lista de vari√°veis para as quais queremos criar tabelas de conting√™ncia
variables = ['V6461_v01', 'V6471_v01', 'V6121_v01', 'V0606', 'V0604', 'V5030',
'V1005', 'V0613', 'V0614', 'V0615', 'V0616', 'V0617', 'V0618',
'V0619', 'V0627', 'V0637', 'V0640', 'V0648', 'V0650', 'V0656',
'V0657', 'V0658', 'V0659', 'V0662', 'V6930', 'V5040', 'V6940',
'V6511_v01_Faixas', 'V6513_v01_Faixas', 'V0653_v01_Faixas',
'V0663', 'V5110', 'V0639', 'V0601', 'V1006', 'V6400','V6036_v01_Faixas', 'V5060_v01_Faixas']

# Lista para armazenar os DataFrames individuais
dataframes = []

# Loop sobre cada vari√°vel
for var in variables:
    # Criar a tabela de conting√™ncia
    contingency_table = pd.crosstab(df_pessoas_v01[var], df_pessoas_v01['Cluster_v01'], dropna=False)

    # Renomear as colunas combinando o nome da vari√°vel com a categoria
    new_column_names = [f'Cluster_{col}' for col in contingency_table.columns]
    contingency_table.columns = new_column_names

    # Transformar a tabela de conting√™ncia em um DataFrame
    df_contingency = contingency_table.reset_index()

    # Renomear a coluna de categorias
    df_contingency.rename(columns={var: 'Categorias'}, inplace=True)

    # Adicionar a coluna "Vari√°vel" com o nome da vari√°vel repetido
    df_contingency['Vari√°vel'] = var

    # Reorganizar as colunas para ter "Vari√°vel" primeiro
    df_contingency = df_contingency[['Vari√°vel', 'Categorias'] + new_column_names]

    # Adicionar o DataFrame individual √† lista
    dataframes.append(df_contingency)

# Concatenar todos os DataFrames individuais em um √∫nico DataFrame
final_df = pd.concat(dataframes, ignore_index=True)

# Exibir o DataFrame final
final_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.decomposition import PCA

# Carregar o DataFrame (substitua pelo seu m√©todo de carregamento)
# df_pessoas_v01_Rio_de_Janeiro_v02 = pd.read_csv('caminho/para/o/seu/arquivo.csv')

# Selecionar todas as 127 colunas para clusteriza√ß√£o
X = df_pessoas_v01_Rio_de_Janeiro_v02

# Definir o range de n√∫meros de clusters a serem testados
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9]

# Loop para avaliar diferentes n√∫meros de clusters
for n_clusters in range_n_clusters:
    # Criar um sub-plot com 1 linha e 2 colunas
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # O 1¬∫ subplot √© o plot de silhueta
    ax1.set_xlim([-0.1, 1])
    # (n_clusters+1)*10 √© para inserir espa√ßo entre os plots de silhueta
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Inicializar o clusterizador KMeans com o n√∫mero de clusters atual
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # A pontua√ß√£o de silhueta m√©dia para todos os amostras
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"Para n_clusters = {n_clusters}, a pontua√ß√£o m√©dia de silhueta √© : {silhouette_avg}")

    # Calcular as pontua√ß√µes de silhueta para cada amostra
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Agregar os scores de silhueta para as amostras pertencentes ao cluster i
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Rotular os plots de silhueta com os n√∫meros dos clusters no meio
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Calcular o novo y_lower para o pr√≥ximo plot
        y_lower = y_upper + 10  # 10 para o espa√ßo entre clusters

    ax1.set_title("Plot de Silhueta para v√°rios clusters")
    ax1.set_xlabel("Valores de coeficiente de silhueta")
    ax1.set_ylabel("R√≥tulo do Cluster")

    # Linha vertical para a pontua√ß√£o m√©dia de silhueta de todos os valores
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Limpar os r√≥tulos / ticks do eixo y
    ax1.set_xticks([i / 10.0 for i in range(-1, 11)])

    # 2¬∫ plot mostrando os clusters formados
    # Para visualizar em 2D, usamos PCA para reduzir a dimensionalidade
    pca = PCA(n_components=2)
    X_reduced = pca.fit_transform(X)

    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X_reduced[:, 0], X_reduced[:, 1], marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Rotular os clusters
    centers = clusterer.cluster_centers_
    centers_reduced = pca.transform(centers)

    ax2.scatter(centers_reduced[:, 0], centers_reduced[:, 1], marker='o', c='white', alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers_reduced):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50, edgecolor='k')

    ax2.set_title("Visualiza√ß√£o dos dados clusterizados")
    ax2.set_xlabel("Espa√ßo das features reduzido para 2D - PCA")
    ax2.set_ylabel("Espa√ßo das features reduzido para 2D - PCA")

    plt.suptitle(f"An√°lise de Silhueta para clustering KMeans com n_clusters = {n_clusters}",
                 fontsize=14, fontweight='bold')

    plt.show()
```
